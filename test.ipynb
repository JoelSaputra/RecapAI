```python
import os 
from dotenv import load_dotenv
from google import genai 


load_dotenv()


client = genai.Client(api_key=os.environ["GEMINI_API_KEY"])




def summarize_news():
    headline = "Carlsberg reported to have filed draft papers for IPO to raise $700 million"
    url = "https://www.bloomberg.com/news/articles/2026-07-02/carlsberg-said-to-file-for-india-ipo-to-raise-700-million"
    interaction = client.interactions.create(
    model="gemini-3.5-flash",
    input=f"Summarize this news in 4 sentences:\n\nHeadline: {headline}\nURL: {url}"
)
    
    print(interaction.output_text)
```
